```python
# Load number counts

try:
    import cpickle as pickle
except ModuleNotFoundError:
    import pickle
def open_object(filename):
    with open(filename, 'rb') as inp:
        obj = pickle.load(inp)
    return obj

ncounts = open_object(path_to_experiment_storage/"number_counts_per_epoch.pkl")
ncounts
```

```python
# ncounts is a dictionary of epcohs:
ncounts.keys()
```

```python
# Look at epoch0
epoch0_ncounts = ncounts[0]
epoch0_ncounts
```

```python
e0l2 = epoch0_ncounts[epoch0_ncounts["layer_idx"] == 2]
e0l2
```

```python
e0l3 = epoch0_ncounts[epoch0_ncounts["layer_idx"] == 3]
e0l3
```

```python
e0l4 = epoch0_ncounts[epoch0_ncounts["layer_idx"] == 4]
e0l4
```

```python
# SANITY CHECK: The total number of counts for each layer should be the same. 

for layer_counts in [e0l1, e0l2, e0l3, e0l4]:
    total_counts = layer_counts["total"].sum()
    print(f"Total counts for layer {layer_counts['layer_idx'].iloc[0]}: {total_counts}")
```
