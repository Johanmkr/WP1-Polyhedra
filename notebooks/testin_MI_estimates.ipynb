```python
# For each epoch, compute the layer-wise quantites and store them in a dictionary. 

# First, make sure the counts are sorted appropriately. 


epoch0_ncounts
```

```python
# Make sure the two first columns are layer_idx and region_idx

epoch0_ncounts = epoch0_ncounts.sort_values(by=["layer_idx", "region_idx"]).reset_index(drop=True)
epoch0_ncounts  
```

```python
def check_number_counts_consistency(ncounts):
    # Check if last column is "total" and check if the sum of total counts is the same for each layer.
    assert ncounts.columns[ -1 ] == "total", "Last column should be 'total'"

    # Find the class columns as the third to the second last columns
    class_columns = ncounts.columns[ 2 : -1 ]
    class_columns_list = class_columns.tolist()

    # Check that the horisontal sum of classes equals the total counts for each row
    # Find sum of all classes in one single row (region)
    row_sums = ncounts[class_columns_list].sum(axis=1)
    totals = ncounts["total"]
    # Check if all row sums equal total counts
    assert all(row_sums == totals), "Row sums of class counts should equal total counts"

    # Find number of unique layers
    num_layers = ncounts["layer_idx"].nunique()
    # Get total counts per layer
    total_counts_per_layer = ncounts.groupby("layer_idx")["total"].sum().values
    # Check if all total counts are the same
    assert all(total_counts_per_layer == total_counts_per_layer[0]), "Total counts should be the same for each layer"
    
    return ncounts, num_layers, class_columns_list
    
ncounts0, num_layers, classes = check_number_counts_consistency(epoch0_ncounts)
```

```python
ncounts0
```

```python
# Find the data per class and check that they match for all layers. 

# This should go in the consistency check
for layer in range(1, num_layers+1):
    layer_counts = ncounts0[ncounts0.layer_idx == layer]
    data_per_class = {classi: [np.sum(layer_counts[classi])] for classi in classes}
```

```python
layer_counts = ncounts0.loc[ncounts0["layer_idx"] == 1, classes]

n_k = layer_counts.sum(axis=0).to_numpy()[None, :]
n_k
```

```python


ncounts0["KL"] = 0.0
ncounts0["KL2"] = 0.0
# Find n_w - total points per region
n_w = np.array(ncounts0["total"])

# Find n_kw - total points per region per class
n_kw = np.array(ncounts0[classes])

# Find n_k - total points per class
# n_k = np.array(pd.DataFrame.from_dict(data_per_class)[classes])

N = n_k.sum()

# Normalise to make prob mass
m_w = np.expand_dims(n_w / N, axis=1) 
m_kw = n_kw / N
m_k = n_k / N



for cl in classes:
    n_kw = np.array(ncounts0[cl])
    n_k = np.array(data_per_class[cl])
    
    # Two-step logarithm to avoid 0 in log
    logterm = N * n_kw / (n_k*n_w)
    logterm = np.log(logterm, where=logterm > 0)
    

    ncounts0["KL"] += n_kw / N * logterm
    
ncounts0
```

```python
print(f"m_w shape: {m_w.shape}")
print(f"m_kw shape: {m_kw.shape}")
print(f"m_k shape: {m_k.shape}")


logterm = m_kw / (m_w @ m_k)
logterm = np.log(logterm, where=logterm > 0)

ncounts0["KL2"] = (m_kw * logterm).sum(axis=1)
```

```python
ncounts0
```

```python
# Estimate layer-wise MI for epoch0
MI_KL_per_layer = {}
for layer in range(1,num_layers+1):
    MI_KL_per_layer[layer] = sum(ncounts0[ncounts0.layer_idx==layer]["KL"])
    
MI_KL_per_layer
```
