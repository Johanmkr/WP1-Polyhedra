```python
import os, sys
sys.path.append(os.path.abspath('src'))
from utils import animate_training
```

```python
T1 = trainer.MoonTrainer(models.MLP21())
T2 = trainer.MoonTrainer(models.MLP31())
T3 = trainer.MoonTrainer(models.MLP231())
T4 = trainer.MoonTrainer(models.MLP321())


EPOCHS = 5000
```

```python
T1.train(epochs=EPOCHS)
T2.train(epochs=EPOCHS)
T3.train(epochs=EPOCHS)
T4.train(epochs=EPOCHS)
```

```python
#% matplotlib inline
# Plot losses
fig, ax = plt.subplots(figsize=(10, 7))
for T in [T1, T2, T3, T4]:
    ax.plot(T.losses, label=T.model.name)

ax.set_title('Losses')
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.legend()
plt.show()
```
