## What are the regions?

Any CPWL neural network divide an $n_0$-dimensional input space into a partition $\Omega^{n_0}$ which is the disjoint union of individual regions $\omega$ such that:
$$
\Omega^{n_0} = \left\{\omega: \bigcup_i\omega_i = \mathbb{R}^{n_0},\; \omega_i\cap\omega_j = \emptyset \;\forall \;i,j \right\}.
$$

These regions are formed by a set of hyperplanes $\mathcal{H}$ in input space and can be represented as:

* $\mathcal{H}$-representation: Solution to a set of inequalitites:
$$
    \omega = \{\mathbf{x}\in\mathbb{R}^{n_0}: D_\omega\mathbf{x}\leq\mathbf{g}_\omega \}
$$
* $\mathcal{V}$-representetion: $\omega = \textrm{Conv}(\mathcal{V})$. 

## How do we find the hyperplanes?

### TL;DR
Any activation pattern $\mathbf{q}_\omega \to \mathbf{s}_\omega \; \cup( W, \mathbf{b}) \to D_\omega, \mathbf{g}_\omega$

Each neuron induce a hyperplane _in its input domain_. Hence, each layer produces a set of hyperplanes. 
\begin{equation*}
    \Omega^{\leq l} = \bigcap_{j=1}^{l} \Omega^j.
\end{equation*}

The whole network is a _piecewise linear affine transformation_:
\begin{equation*}
    F^l_\omega(\mathbf{x}) = A^l_\omega \mathbf{x} + \mathbf{c}^l_\omega, \quad \text{for } \mathbf{x} \in \omega \subset \mathbb{R}^d,
\end{equation*}

with $A^l_\omega$ and $\mathbf{c}^l_\omega$ being recursive projections of weights and biases up to layer $l$. They can be represented fully as:



### Recursive projections

\begin{equation*}
    A^l_\omega = \prod_{i=1}^{l} Q^i_\omega W^i,
\end{equation*}
and
\begin{equation*}
    \mathbf{c}^l_\omega = Q^l_\omega \mathbf{b}^l + \sum_{j=1}^{l-1}\left( \prod_{k=j+1}^{l} Q^k_\omega W^k \right) Q^j_\omega \mathbf{b}^j,
\end{equation*}
where $Q^i_\omega = \mathrm{diag}(\mathbf{q}^i_\omega)$ is a diagonal matrix indicating the activation status $\mathbf{q}^i_\omega$ of neurons in layer $i$ for region $\omega$, $W^i$ and $\mathbf{b}^i$ are the weight matrix and bias vector for layer $i$, respectively.

This also induce the following recursive relations:
\begin{equation*}
    A^l_\omega = Q^l_\omega W^l A^{l-1}_\omega,
\end{equation*}
and
\begin{equation*}
    \mathbf{c}^l_\omega = Q^l_\omega \left( W^l \mathbf{c}^{l-1}_\omega + \mathbf{b}^l \right),
\end{equation*}
with base case $A^0_\omega = I$ and $\mathbf{c}^0_\omega = \mathbf{0}$. The quantity $W^lA^{l-1}_\omega$ may be interpreted as the effective weights of layer $l$ when restricted to region $\omega$. Similarly, $W^l \mathbf{c}^{l-1}_\omega + \mathbf{b}^l$ represents the effective bias of layer $l$ in region $\omega$



### Inequalitites
Find a sign matrix $S$ from $Q$ that maps $(0,1)\to(1,-1)$:
\begin{equation*}
    \mathbf{s}^l_\omega = -2\mathbf{q}^l_\omega + 1,
\end{equation*}
so that we can write:
\begin{equation*}
    S^l_\omega \left( W^l A^{l-1}_\omega \mathbf{x} + W^l \mathbf{c}^{l-1}_\omega + \mathbf{b}^l \right) \geq 0.
\end{equation*}

By aggregating the inequalities from all layers up to $l$, we can fully characterize the convex region $\omega$ in the input space.

For simplicity, we define $D^l_\omega = S^l_\omega W^lA^{l-1}_\omega$ and $\mathbf{g}^l_\omega = -S^l_\omega (W^l \mathbf{c}^{l-1}_\omega + \mathbf{b}^l)$, allowing us to express the inequalities defining region $\omega$ as:
\begin{equation*}
    \omega = \{\mathbf{x} \in \mathbb{R}^{n_0} : D^l_\omega \mathbf{x} \leq \mathbf{g}^l_\omega\}.
\end{equation*}

## Original idea: Use these partitions for MI - estimation

Why? Because analysing $I(X;T)$ and $I(Y;T)$ is widely used in literature. Affine transformations are deterministic: $T = AX + b$ for some $A$ and $b$. Hence: 

$$I(Y;T)=I(Y;\Omega) + I(Y;X|\Omega) $$ $$ I(X;T)=I(X;\Omega) + I(X;X|\Omega)$$

We look at $I(Y;\Omega)$:
$$
    I(Y;\Omega) = \sum_{\omega\in\Omega}\sum_{k\in Y}\frac{n_{k\omega}}{N}\log{\frac{n_{k\omega}N}{n_\omega n_k}},
$$
where
* $N$ is the total number of points.
* $n_{k\omega}$ is the number of points of class $k$ in region $\omega$.
* $n_k$ is the total number of points of class $k$.
* $n_\omega$ is the total number of points in region $\omega$.

## MI $\to$ Network Capacity / Expressivity

### Theoretical bounds
Denote the number of regions: $R$. Naïve bounds are: $R_\mathrm{min} \geq 1$ and $R_\mathrm{max}\leq 2^H$ for a total of $H$ neurons. 

#### Shallow networks
One-layer networks are bounded by Zaslavsky's theorem:
$$
    R_\mathrm{max} \leq \sum_{i=0}^{n_0}\binom{n}{i}
$$
How many regions can an arrangement of $n$ hyperplanes of dimension $n_0$ create (in general position)? This scales as $\mathcal{O}(n^{n_0})$ (polynomial) with parameters $\mathcal{O}(n)$

#### Deep networks - account for space folding
For a deep network with $L$ layers and hidden neuron config: [$n_1, n_2, \dots n_L$] the maximum number of regions is lower bounded by:
$$
    R_\mathrm{max} \leq \left(   \prod_{i=1}^{L-1}\left\lfloor\frac{n_i}{n_0}\right\rfloor^{n_0} \right)\sum_{j=0}^{n_0}\binom{n_L}{j}.
$$
This scales as $\Omega\left(  \binom{n}{n_0}^{(L-1)n_0}n^{n_0}  \right)$ (combinatorial/exponential) with parameters $\mathcal{O}(Ln^2)$.
