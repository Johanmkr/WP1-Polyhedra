```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_side_by_side_hamming(df_results: pd.DataFrame, title=""):
    """
    Plots Intra- and Inter-class Hamming distances in two side-by-side subplots
    sharing the same Y-axis for easy comparison.
    """
    # Create a figure with 2 subplots side-by-side sharing the Y-axis
    fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
    
    # Use a visually distinct color palette
    palette = sns.color_palette("tab10", as_cmap=False, n_colors=df_results["layer_idx"].nunique())

    # Left Plot: Intra-class
    sns.lineplot(data=df_results, x="epoch", y="avg_intra_hamming", hue="layer_idx", 
                 palette=palette, marker="X", linewidth=2, ax=axes[0])
    axes[0].set_title("Intra-class Hamming Distance\n(Lower is better: classes are tightly clustered)", fontsize=13)
    axes[0].set_xlabel("Epoch", fontsize=12)
    axes[0].set_ylabel("Hamming Distance (Bits)", fontsize=12)
    axes[0].grid(True, linestyle='--', alpha=0.6)

    # Right Plot: Inter-class
    sns.lineplot(data=df_results, x="epoch", y="avg_inter_hamming", hue="layer_idx", 
                 palette=palette, marker="o", linewidth=2, ax=axes[1])
    axes[1].set_title("Inter-class Hamming Distance\n(Higher is better: classes are pushed apart)", fontsize=13)
    axes[1].set_xlabel("Epoch", fontsize=12)
    axes[1].grid(True, linestyle='--', alpha=0.6)

    # Clean up duplicate legends
    axes[0].get_legend().remove()
    axes[1].legend(title="Layer Index", bbox_to_anchor=(1.05, 1), loc='upper left')
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

# Usage:
plot_side_by_side_hamming(res)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def compute_metric_correlations(df_results: pd.DataFrame) -> pd.DataFrame:
    """
    Computes the correlation between structural Hamming distances 
    and Mutual Information metrics per layer over the training trajectory.
    """
    correlations = []
    
    for layer, group in df_results.groupby("layer_idx"):
        # We need at least a few epochs to compute a meaningful correlation
        if len(group) < 3:
            continue
            
        # 1. Intra-class vs. Memorization/Compression I(X;W)
        # Hypothesis: As the network compresses (I(X;W) drops), classes cluster tighter (Intra drops).
        # We expect a strong POSITIVE correlation.
        corr_intra_xw = group["avg_intra_hamming"].corr(group["I(X;W)"])
        
        # 2. Inter-class vs. Predictive Capacity I(Y;W)
        # Hypothesis: As the network learns to predict (I(Y;W) rises), classes push apart (Inter rises).
        # We expect a strong POSITIVE correlation.
        corr_inter_yw = group["avg_inter_hamming"].corr(group["I(Y;W)"])
        
        correlations.append({
            "layer_idx": layer,
            "corr(Intra, I(X;W))": round(corr_intra_xw, 4),
            "corr(Inter, I(Y;W))": round(corr_inter_yw, 4)
        })
        
    df_corr = pd.DataFrame(correlations)
    return df_corr

def plot_correlation_scatter(df_results: pd.DataFrame, layer_idx: int):
    """
    Visually verifies the correlation for a specific layer using a dual scatter plot.
    """
    layer_data = df_results[df_results["layer_idx"] == layer_idx]
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Left: Compression vs Clustering
    sns.regplot(data=layer_data, x="I(X;W)", y="avg_intra_hamming", ax=axes[0], color="purple")
    axes[0].set_title(f"Layer {layer_idx}: Compression vs. Clustering\nExpect an upward trend")
    
    # Right: Prediction vs Separation
    sns.regplot(data=layer_data, x="I(Y;W)", y="avg_inter_hamming", ax=axes[1], color="teal")
    axes[1].set_title(f"Layer {layer_idx}: Prediction vs. Separation\nExpect an upward trend")
    
    plt.tight_layout()
    plt.show()

# ==========================================
# Usage:
# ==========================================
df_corr = compute_metric_correlations(res_nor)
print(df_corr)
plot_correlation_scatter(res_nor, layer_idx=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_2x2_correlations(df_results: pd.DataFrame):
    """
    Plots a 2x2 grid of the regression lines between the structural metrics 
    (Hamming Distances) and the information metrics (Mutual Information).
    Includes all layers distinguished by color, without scatter points or error bands.
    """
    # Create the 2x2 grid
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Define the configurations for the 4 subplots
    # Tuple format: (X-axis column, Y-axis column, Row idx, Col idx, Plot Title)
    plot_configs = [
        ("I(X;W)", "avg_intra_hamming", 0, 0, "Intra-class vs. Memorization I(X;W)"),
        ("I(Y;W)", "avg_intra_hamming", 0, 1, "Intra-class vs. Prediction I(Y;W)"),
        ("I(X;W)", "avg_inter_hamming", 1, 0, "Inter-class vs. Memorization I(X;W)"),
        ("I(Y;W)", "avg_inter_hamming", 1, 1, "Inter-class vs. Prediction I(Y;W)")
    ]
    
    # Extract unique layers and generate a distinct color palette
    layers = sorted(df_results["layer_idx"].unique())
    palette = sns.color_palette("tab10", n_colors=len(layers))
    
    # Loop through each of the 4 subplots
    for x_col, y_col, row, col, title in plot_configs:
        ax = axes[row, col]
        
        # Loop through each layer to plot its specific regression line
        for i, layer in enumerate(layers):
            layer_data = df_results[df_results["layer_idx"] == layer]
            
            # Plot ONLY the regression line (no scatter, no confidence intervals)
            sns.regplot(
                data=layer_data, 
                x=x_col, 
                y=y_col, 
                scatter=False,    # Removes the data points
                ci=None,          # Removes the uncertainty/confidence bands
                ax=ax, 
                color=palette[i],
                line_kws={"linewidth": 2.5},
                # Only add label on the first plot to avoid duplicate legends
                label=f"Layer {layer}" if (row == 0 and col == 0) else None 
            )
            
        # Format the subplot
        ax.set_title(title, fontsize=14, pad=10)
        ax.set_xlabel(f"{x_col} (Bits)", fontsize=12)
        ax.set_ylabel(y_col.replace('_', ' ').title(), fontsize=12)
        ax.grid(True, linestyle='--', alpha=0.5)

    # --- THE FIX IS HERE ---
    # Extract the legend from the first subplot and place it globally on the figure
    handles, labels = axes[0, 0].get_legend_handles_labels()
    
    fig.legend(handles, labels, title="Network Layer", loc='center right', 
               bbox_to_anchor=(1.08, 0.5), fontsize=12, title_fontsize=13)
    
    # Adjust layout so subplots don't overlap with the external legend
    plt.tight_layout(rect=[0, 0, 0.95, 1])
    plt.show()
```

```python
import pandas as pd

def compute_all_correlations(df_results: pd.DataFrame) -> pd.DataFrame:
    """Calculates Pearson correlation for all four metric pairs per layer."""
    correlations = []
    
    for layer, group in df_results.groupby("layer_idx"):
        if len(group) < 3: # Need a few epochs to calculate correlation
            continue
            
        correlations.append({
            "Layer": layer,
            "Intra vs I(X;W)": group["avg_intra_hamming"].corr(group["I(X;W)"]),
            "Intra vs I(Y;W)": group["avg_intra_hamming"].corr(group["I(Y;W)"]),
            "Inter vs I(X;W)": group["avg_inter_hamming"].corr(group["I(X;W)"]),
            "Inter vs I(Y;W)": group["avg_inter_hamming"].corr(group["I(Y;W)"])
        })
        
    return pd.DataFrame(correlations).set_index("Layer")

# Create the summary dataframe
df_corr = compute_all_correlations(res_nor)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_heatmap(df_corr: pd.DataFrame):
    plt.figure(figsize=(10, 6))
    
    # Plot heatmap with values annotated, fixed scale from -1 to 1
    sns.heatmap(df_corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1, 
                center=0, fmt=".2f", linewidths=0.5)
    
    plt.title("Correlation: Hamming Distances vs. Mutual Information\n(Across Training Epochs)", 
              fontsize=14, pad=15)
    plt.ylabel("Network Layer", fontsize=12)
    plt.xlabel("Metric Pairs", fontsize=12)
    
    plt.tight_layout()
    plt.show()

# Usage:
plot_correlation_heatmap(df_corr)
```

```python
def plot_correlation_depth(df_corr: pd.DataFrame):
    plt.figure(figsize=(10, 6))
    
    # We plot each column (metric pair) as a separate line
    markers = ['o', 's', '^', 'D']
    for idx, column in enumerate(df_corr.columns):
        plt.plot(df_corr.index, df_corr[column], 
                 marker=markers[idx], linewidth=2.5, label=column)
        
    # Add a dashed line at 0 (no correlation) for reference
    plt.axhline(0, color='black', linestyle='--', alpha=0.5)
    
    plt.title("How Correlation Evolves with Network Depth", fontsize=14, pad=15)
    plt.xlabel("Layer Index", fontsize=12)
    plt.ylabel("Pearson Correlation Coefficient (r)", fontsize=12)
    plt.ylim(-1.1, 1.1)
    
    plt.legend(title="Metric Pair", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True, linestyle=':', alpha=0.7)
    
    plt.tight_layout()
    plt.show()

# Usage:
plot_correlation_depth(df_corr)
```
